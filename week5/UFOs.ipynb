```python
import re
inpath = 'ufo_awesome.tsv'
inf = open(inpath, 'r')

for i, line in enumerate(inf):
    splitline = line.split('\\t')
    if len(splitline) != 6:
        first_bad_line = splitline
        print("First bad row:", i)
        for j, col in enumerate(first_bad_line):
            print(j, col)
        break

inf.close()
```

```python
max_count = 20
count = 0

nc = re.compile(r"NC")
city_state = re.compile(r"\s*(?P<city>.+), (?P<state>[A-Z]{2})\s*$")
# city_state = re.compile(r"Durham, NC")
events_by_city = {}

with open(inpath) as f:
    reader = csv.DictReader(f, fieldnames=['date_reported', 'date_occurred', 'location', 'short description', 'duration', 'description'], delimiter="\t")
    for row in reader:
        match = city_state.search(row['location'])
        if match:
            #city, state = match.groups()
            events_by_city.setdefault((match.group('city'), match.group('state')), []).append(row)
#         print(row)
#        if nc.search(row['location']):
#            print(row)
#        else:
#            continue
#        
#        if count > max_count:
#            break
#        count += 1
#        print(row)
```
