```python
class StockData:
    def __init__(self, **kwargs):
        self.date = kwargs["date"]
        self.open = float(kwargs["open"])
        self.high = float(kwargs["high"])
        self.low = float(kwargs["low"])
        self.close = float(kwargs["close"])
        self.volume = float(kwargs["volume"])

        
    def avg(self):
        l = [self.open, self.close, self.low, self.high]
        return sum(l)/len(l)
    
    def __repr__(self):
        return "{} {} {} {} {} {} {}".format(self.date, self.open, 
                                             self.high, self.low, 
                                             self.close, self.volume, 
                                             self.avg())
        
```

```python
data = []
with open('EOD-GOOG.csv') as f:
    reader = csv.DictReader(f)
    for row in reader:
        data.append(StockData(**row))

```

```python
def get_avg(**kwargs):
    return (float(kwargs["open"]) + float(kwargs["close"]) + 
            float(kwargs["high"]) + float(kwargs["low"]))/4
    
with open("EOD-GOOG.csv") as f:
    reader = csv.DictReader(f)
    for row in reader:
        print(row["date"], get_avg(**row))
        
```
